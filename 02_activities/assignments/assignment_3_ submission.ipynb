### <b>1st visualization using Python</b>

```python
# For visualization:

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.express as px
```

```python
# Loading the RentSafeTO: Apartment Building Evaluations data (downloaded as .csv file)
df = pd.read_csv('C:/D Drive/DSI_UofT/dsiVisualization/visualization/02_activities/assignments/city_of_Toronto_Open_Data_Portal_Apartment_Building_Evaluations_2023_current.csv')

# Convert column names to snake case
df.columns = df.columns.str.lower().str.replace(' ', '_')

# ensuring that the values are numbers and drop any missing values
df['current_building_eval_score'] = pd.to_numeric(df['current_building_eval_score'], errors='coerce')
df_clean = df.dropna(subset=['current_building_eval_score', 'wardname'])
print(df.columns)
```

```python
# Next sort by median score for better visual hierarchy
ward_order = df_clean.groupby('wardname')['current_building_eval_score'].median().sort_values().index
```

```python
# Next steps for data visualization
fig = px.box(
    df_clean,
    y='wardname',
    x='current_building_eval_score',
    orientation='h', # Horizontal plot
    title='Distribution of RentSafeTO Building Scores by Ward (Sorted by Median)',
    category_orders={'wardname': ward_order}, # Apply the sorted order
    color='wardname', # Color by ward name
    color_discrete_sequence=px.colors.qualitative.Safe, # color-blind friendly discrete sequence
    hover_data={'current_building_eval_score': True, 'wardname': False} # Customize hover info
)
# Add the vertical line for the minimum passing score
fig.add_vline(x=50, line_width=2, line_dash="dash", line_color="black", annotation_text="Minimum Passing Score", annotation_position="bottom right")

# Update layout for better aesthetics
fig.update_layout(
    xaxis_title='Evaluation Score (0-100)',
    yaxis_title='Toronto Ward Name',
    showlegend=False # Hide the legend as colors are clear from y-axis
)

# Display the interactive figure
fig.show()
```
