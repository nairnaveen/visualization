```python
# loading libraries
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

## Making a basic figure with matplotlib

```python
# make some sample data so that we have something to plot!
np.random.seed(613)
x = np.arange(50)
y = np.random.randint(0,100,50)

# make our basic scatterplot
fig, ax = plt.subplots(figsize=(5,3))
ax.scatter(x,y)
```

```python
# modifying our graph; keep our figure and axis definition the same and only edit our axis object
np.random.seed(613)
x = np.arange(50)
y = np.random.randint(0,100,50)

fig, ax = plt.subplots(figsize=(5,3))
ax.bar(x,y)
```

```python
# modifying our graph
np.random.seed(613)
x = np.arange(50)
y = np.random.randint(0,100,50)

fig, ax = plt.subplots(figsize=(5,3))
ax.barh(x,y)
```

```python
# make a line plot
np.random.seed(613)
x = np.arange(50)
y = np.random.randint(0,100,50)

fig, ax = plt.subplots(figsize=(5,3))
ax.plot(x,y)
```

```python
# modifying our graph
np.random.seed(613)
x = np.arange(50)
y = np.random.randint(0,100,50)

fig, ax = plt.subplots(figsize=(5,3))
ax.hist(y)
```

```python
# Adding axis labels and titles
fig, ax = plt.subplots(figsize=(5,3))
ax.plot(x,y)
ax.set_title('Total growth over time')
ax.set_ylabel('Total growth')
ax.set_xlabel('Years since start')
fig.tight_layout()
```

```python
# modify our labels and titles’ appearance by using font dictionaries (‘fontdict’)
font1 ={'family':'sans-serif','color':'blue', 'size':20} # dictionary with the properties we want for our labels
font2 ={'family':'monospace','color':'green', 'size':14}

fig, ax = plt.subplots(figsize=(5,3))
ax.plot(x,y)
ax.set_title('Total growth over time', fontdict=font1)
ax.set_ylabel('Total growth', fontdict=font2)
ax.set_xlabel('Years since start', fontdict=font2)
fig.tight_layout()
```

```python
# add fontdict arguments to our previous lineplot
font1 ={'family':'sans-serif','color':'blue', 'size':20}
font2 ={'family':'monospace','color':'green', 'size':14}

fig, ax = plt.subplots(figsize=(5,3))
ax.plot(x,y)
ax.set_title('Total growth over time', fontdict=font1, loc = 'left')
ax.set_ylabel('Total growth', fontdict=font2)
ax.set_xlabel('Years since start', fontdict=font2)
fig.tight_layout()
```

```python
# Moving labels and titles
# move our lables with the loc argument
fig, ax = plt.subplots(figsize=(5, 3))
ax.plot(x,y)
ax.set_ylabel('Total growth', fontdict = font2)
ax.set_xlabel('Years since start', fontdict = font2)
```

```python
# add color and marker arguments (NOTE:American spelling)
fig, ax = plt.subplots(figsize=(5,3))
ax.scatter(x,
           y,
           marker = '*', # NEW ADDITIONS
           color = 'indigo') # NEW ADDITIONS
```

```python
fig, ax = plt.subplots(figsize=(5,3))
ax.scatter(x,y,marker = '+',color = 'indigo')
```

```python
fig, ax = plt.subplots(figsize=(5,3))
ax.scatter(x,y,marker = 'H',color = 'indigo')
```

```python
fig, ax = plt.subplots(figsize=(5,3))
ax.plot(x,y,marker = '*',color = 'indigo', linestyle = '--', linewidth = 2)
```

```python
# modify colour using named colors (https://matplotlib.org/stable/gallery/color/named_colors.html#sphx-glr-gallery-color-named-colors-py)
fig, ax = plt.subplots(figsize=(5,3))
ax.plot(x,
        y,
        marker = '*',
        color = '#7425b9', # use hex codes
        linestyle = '--', 
        linewidth = 2)
```

```python
# customizing plot further
fig, ax = plt.subplots(figsize=(5,3))
ax.plot(x,y,
        marker = '*',
        markersize = 12, # NEW ADDITIONS
        color = '#7425b9', 
        linestyle = '--', 
        linewidth = 2,
        markeredgecolor = '#fa9359', # NEW ADDITIONS
        markerfacecolor = '#000000') # NEW ADDITIONS
```

```python
# add grid lines to our axis object using .grid()
fig, ax = plt.subplots(figsize=(5,3))
ax.plot(x,y,
        marker = '*',
        markersize = 12,
        color = '#7425b9', 
        linestyle = '--', 
        linewidth = 2,
        markeredgecolor = '#fa9359',
        markerfacecolor = '#000000')
ax.grid(axis = 'y')
```

```python
fig, ax = plt.subplots(figsize=(5,3))
ax.plot(x,y,
        marker = '*',
        markersize = 12,
        color = '#7425b9', 
        linestyle = '--', 
        linewidth = 2,
        markeredgecolor = '#fa9359',
        markerfacecolor = '#000000')
ax.grid(color = 'blue',  # modify the appearance (colour, width, linestyle) of our grid lines
        linestyle = 'dotted', 
        linewidth = 2)
```
