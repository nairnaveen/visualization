```python
# import libraries
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# dataset used is the iris dataset loaded using seaborn.
df = sns.load_dataset('iris')
```

```python
# call the kdeplot() function again to add another element to the plot.
sns.set_theme(style="darkgrid")
sns.kdeplot(df['sepal_width'], fill=True, color="r")
sns.kdeplot(df['sepal_length'], fill=True, color="b")
plt.show()
```

```python
# flipped the plot by 90 degrees orientation thanks to the y argument.
sns.set_theme(style="darkgrid")
sns.kdeplot(y=df['sepal_width'], fill=True, color="r")
sns.kdeplot(y=df['sepal_length'], fill=True, color="b")
plt.show()
```
